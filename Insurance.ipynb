```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.info()
```

```python
null_cols = df.columns[df[df == -1].any()]
print (df[null_cols] == -1).sum()
```

```python
df_test = pd.read_csv('test.csv')
result = pd.DataFrame(df_test['id'])
```

```python
train = df.drop(null_cols, axis=1)
test = df_test.drop(null_cols, axis=1)
y = train['target']
train.drop(['id', 'target'], axis=1, inplace=True)
test.drop('id', axis=1, inplace=True)
X = train.as_matrix().astype(np.float)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
rfc = RandomForestClassifier(n_estimators = 100,min_samples_leaf = 10,min_samples_split=15)
rfc.fit(X_train, y_train)
print accuracy_score(y_test, rfc.predict(X_test))
```

```python
def calcginiindex(array):

    array = array.flatten()
    array += 0.0000001
    array = np.sort(array)
    index = np.arange(1,array.shape[0]+1)
    n = array.shape[0]
    return ((np.sum((2*index - n - 1)*array))/(n * np.sum(array)))
y_pred = rfc.predict_proba(test)
print calcginiindex(y_pred[:, 1])
```

```python
result = result.assign(target=y_pred[:, 1])
result.to_csv('randomforest.csv', index=False)
```
